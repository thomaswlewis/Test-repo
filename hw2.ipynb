```python
#problem 2
class mydata:
    def __init__(self, lst):
        self.l = lst
    
    def mean(self):
        return sum(self.l)/len(self.l)
    
    def std(self):
        avg = sum(self.l)/len(self.l)
        sum_sq_dffs = sum((x - avg)**2 for x in self.l)
        return (sum_sq_dffs/len(self.l))**(1/2)

x = mydata([1, 2, 7, 5])
print("mean of x =", x.mean())
print("standard dev of x =", x.std())
```

```python
#problem 3
import numpy as np

def lag(vectr,int_lag):
    lngth = len(vectr)
    end = lngth - int_lag
    return np.concatenate((np.ones(int_lag)*np.nan, vectr[:end]))
    
print(lag(np.array( [1, 6, 8, 4] ),1))
```

```python
#problem 4
def lagGenerator(n):
    return lambda x:lag(x,n)

lag2 = lagGenerator(2)
print(lag2(np.array( [1, 6, 8, 4] )))
```
